```python
import pandas as pd
import yaml
```

```python
schema_dir = "schema.yml"
rename_cols_dir = "rename_cols.yml"
```

```python
url_csv = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/yellow_tripdata_2019-01.csv.gz"

df_csv = pd.read_csv(url_csv, compression="gzip")
df_csv.head()
```

```python
df_csv.info()
```

```python
url_pq = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2019-01.parquet"

df_pq = pd.read_parquet(url_pq, engine="pyarrow")
df_pq.head()
```

```python
df_pq.info()
```

```python
service = "yellow"
url_pq = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2019-01.parquet"

with open(schema_dir, 'rb') as f:
  service_schema = yaml.safe_load(f)[service]

with open(rename_cols_dir, "rb") as f:
  service_rename_cols = yaml.safe_load(f)[service]

df_pq = pd.read_parquet(url_pq, engine="pyarrow").astype(service_schema)
df_pq = df_pq.rename(columns=service_rename_cols)
df_pq.info()
```
